# NumPy 2-D Arrays: operations involving axis

```python
import numpy as np
np_array_1d = np.arange(0,6)
print("Working with 1d array")
print(np_array_1d)
print(np_array_1d.shape)
print(np.sum(np_array_1d))
print("Working with 2d arrays")
np_array_2d = np_array_1d.reshape([3,2])
print(np_array_2d)
print(np_array_2d.shape)
print(np.sum(np_array_1d))
print(np_array_2d.flatten())
print("Sum 2d along different axis")
print(np.sum(np_array_2d,axis=0))
print(np.sum(np_array_2d,axis=1))
print(np.sum(np_array_2d))
```

# In-class activity 3: Statistics using Numpy 

```python
#Calculate mean and median over all elements of a 2d numpy array with and without calling flatten() 
b = np.arange(12).reshape((2, 6))
print(b)
print(np.median(b))
print(np.mean(b))
print(type(b.flatten()))
print(b.flatten())
print(np.mean(b.flatten()))
print(b.flatten().mean())
print(np.median(b.flatten()))
```

```python
#Calculate the mean, standard deviation, and variance along the second axis
print(b)
print(np.mean(b,axis=1))
print(np.average(b,axis=1))
print(np.std(b,axis=1))
print(np.var(b,axis=1))
print(np.sqrt(np.var(b,axis=1)))
```

```python
#Compute covariance matrix of two vectors
print(b)
print(np.cov(b))
```

```python
x = np.array([[0, 2, 3], [1, 1, 4], [2, 0, 5]]).T
print(x)
np.cov(x)
```

```python
x1=np.array([0, 1, 2])
x2=np.array([2, 1, 0])
print(x2.shape,type(x2))
x3=np.array([3, 4, 5])
print(np.cov(x1))
print(np.cov(x2))
print(np.cov(x3))
print(np.cov(x1,x2))
print(np.var(x1,ddof=1))
print(np.var(x1))
stack_mat = np.stack((x1,x2),axis=0)
print(stack_mat.shape,stack_mat)
print(np.cov(stack_mat))
stack_mat = np.stack((x1,x2),axis=1)
print(stack_mat.shape,stack_mat)
print(np.cov(stack_mat))
```

```python
#Compute weighted average along the specified axis of a given flattened array
import numpy as np
a = np.arange(9).reshape((3,3))
print("Original flattened array:")
print(a)
print("Weighted average along the specified axis of the above flattened array:")
print(np.average(a, axis=0, weights=[1./4, 1./4, 1]))
```

# Introduction to SciPy 


```python
#cuberoot in special package
from scipy.special import cbrt
#Find cubic root of 512 & 64 using cbrt() function
cb = cbrt([512, 64])
#print value of cb
print(cb)
```

```python
#Exponential function in special package
#Exponential function computes the 10**x element-wise.
from scipy.special import exp10
#define exp10 function and pass value in its
exp = exp10(np.linspace(2,5,num=4))
print(exp)
```

```python
#determinant using linear algebra package 
from scipy import linalg
#define square matrix
two_d_array = np.array([ [4,5], [3,2] ])
#pass values to det() function
linalg.det( two_d_array )
```

```python
#inverse of a matrix using linear algebra package
# define square matrix
two_d_array = np.array([ [4,5], [3,2] ])
#pass value to function inv()
linalg.inv( two_d_array )
```

```python
#Rotation of a image using ndimage (multidimensional image processing package)
from scipy import ndimage, misc
panda = misc.face()
#rotatation function of scipy for image  image rotated 135 degree
panda_rotate = ndimage.rotate(panda, 135)
plt.imshow(panda_rotate)
plt.show()
```

```python
#Numerical integration using integrate
#import square root function from math lib
from scipy import integrate
from math import sqrt
# set  fuction f(x)
f = lambda x, y : 64 *x*y
# lower limit of second integral
p = lambda x : 0
# upper limit of second integral
q = lambda y : sqrt(1 - 2*y**2)
# perform double integration
integration = integrate.dblquad(f , 0 , 2/4,  p, q)
print(integration) 
```
